```python
"cargamos VinosPaises.csv"
paises = pd.read_csv("/content/drive/MyDrive/henry/practicaVinos/VinosPaises.csv") # cargar el archivo
paises= pd.DataFrame(paises) #crea el df
paises['country_id'] = range(len(paises))# agrego un listado numerico
paises.head()
```

```python
# prompt: cambiar en df['country'] por el valor de paises['country_id']


# Merge the dataframes based on the 'country' column
df = pd.merge(df, paises, on='country', how='left')
# Replace 'country' column with 'country_id'
df = df.drop('country', axis=1)
df = df.rename(columns={'country_id': 'country'})

# Display the updated dataframe (optional)
df.head()

```
